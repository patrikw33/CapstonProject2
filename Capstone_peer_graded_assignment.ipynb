## Load Dataset and transform it to dataframe

```python
# get website url
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
# get page url
request_url = requests.get(url)
#Store the contents of the website
p_content = xht.fromstring(request_url.content)
#Parse data that are stored table row (tr)
tr_selector =p_content.xpath('//tr') # table row element
```

```python
#For each row, store each first element (header) and an empty list
col_list=[]
cl=0 # number of column
for el in tr_selector[0]:
    cl+=1
    name=el.text_content()   
    col_list.append((name,[]))

#loop the rows in tr selector from second row
for r in range(1,len(tr_selector)):  
    t_row=tr_selector[r]    
  # the target table has tree columns so the length of each must be 3.
    if len(t_row)!=3:
        break   
   
    cl=0 # number of column
    
   # Get dat from stored in each td (table data)
    for row in t_row.iterchildren():
        td_value = row.text_content() 
        #Check if row is empty
       
       #append data to col_list
        col_list[cl][1].append( td_value)       
        cl+=1
```

```python
col_dict={thead:col for (thead,col) in col_list}
df=pd.DataFrame(col_dict)
df.head(10)
```

```python
df.columns # the name of column need to be changed
```

```python
# rename the columns and lean by removing the 2 first characters (\n)
df.columns= ['Postcode', 'Borough', 'Neighbourhood']
df['Neighbourhood'] = df['Neighbourhood'].str.slice(0,-1)
df.head()
```

```python
# sort the data by PostCode and Borough
df_clean_2  = df_clean_1 .sort_values(by=['Postcode','Borough'])

df_clean_2.head()
```
